
### Q1: What is Matplotlib? Why is it used? Name five plots.
- **Matplotlib** is a popular data visualization library in Python used to create static, animated, and interactive visualizations. 
- It is primarily used for plotting graphs and charts to better understand data patterns and relationships.
- **Five plots using Pyplot:**
  1. Line plot
  2. Bar plot
  3. Scatter plot
  4. Histogram
  5. Pie chart





### Q2: Scatter Plot


```python

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y, color='purple')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```



### Q3: Subplot Function


```python

import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Line 4")

plt.tight_layout()
plt.show()

```


### Q4: Bar Plot


```python

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.bar(company, profit, color='orange')
plt.title('Company Profit')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

# Horizontal Bar Plot
plt.barh(company, profit, color='skyblue')
plt.title('Company Profit (Horizontal)')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()


### Q5: Box Plot
```

```python

import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.show()
```
